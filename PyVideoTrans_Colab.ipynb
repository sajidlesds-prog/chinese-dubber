## Configuration Options

Edit `cfg.json` to change settings:
- `edge_tts_emotion_enabled`: true/false (emotional dubbing)
- `blur_subtitle_area`: true/false (blur hardcoded subs in dramas)
- `blur_subtitle_auto`: true (auto-detect bottom 15%) or false (manual coordinates)
- `blur_subtitle_x`, `blur_subtitle_y`: Manual blur region position
- `blur_subtitle_width`, `blur_subtitle_height`: Manual blur region size
- `tts_type`: 0 = Edge TTS (free)
- `voice_role`: TTS voice (see list_edge_voices output)
- `model_name`: whisper model (large-v3-turbo recommended)

## Troubleshooting

If subtitles are not generated:
1. Check `logs/` directory for error messages
2. Ensure video has clear speech audio
3. Try a different Whisper model (large-v3 instead of large-v3-turbo)

If emotional dubbing doesn't work:
1. Verify `edge_tts_emotion_enabled: true` in cfg.json
2. Check that target language is 'zh-cn' or 'en'
3. View logs: `!cat logs/*.log`

If subtitle blur doesn't work:
1. Verify `blur_subtitle_area: true` in cfg.json
2. For manual region, set x, y, width, height
3. For auto-detect, set `blur_subtitle_auto: true`
