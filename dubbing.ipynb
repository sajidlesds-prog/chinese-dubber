```python
# Create folders and move uploaded file
import os
import shutil
os.makedirs("/content/input_videos", exist_ok=True)
os.makedirs("/content/dubbed_output", exist_ok=True)
os.makedirs("/content/temp", exist_ok=True)

# Move uploaded file
for f in os.listdir("/content"):
    if f.endswith(".mp4"):
        shutil.move(f"/content/{f}", f"/content/input_videos/{f}")
        print(f"Moved {f}")

print("Files:", os.listdir("/content/input_videos"))
```
