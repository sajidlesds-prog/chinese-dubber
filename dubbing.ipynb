## Upload your .mp4 file

Click the folder icon on the left → Upload your .mp4 file
Then move it to /content/input_videos/

```python
# Upload movie & create folders
import os
import shutil
os.makedirs("/content/input_videos", exist_ok=True)
os.makedirs("/content/dubbed_output", exist_ok=True)

# If you uploaded a file, move it to input_videos
for f in os.listdir("/content"):
    if f.endswith(".mp4"):
        shutil.move(f"/content/{f}", f"/content/input_videos/{f}")
        print(f"Moved {f} to input_videos")

print("\nFiles in input_videos:", os.listdir("/content/input_videos"))
```
